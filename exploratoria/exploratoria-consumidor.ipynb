### Transforma as colunas lat long em uma coluna só do tipo geometry

```python
geometry = [Point(xy) for xy in zip(df_novo.longitude, df_novo.latitude)]
crs = {'init': 'epsg:4326'}
gdf_clientes = gpd.GeoDataFrame(df_novo, crs=crs, geometry=geometry)
```

```python
gdf_clientes.head()
```

```python
gdf_setores['qtd_clientes'] = 0
for index,row in gdf_setores.iterrows():
    oi = gdf_clientes[gdf_clientes.geometry.within(row.geometry)]
    if oi.empty == False:
        gdf_setores.at[index,'qtd_clientes'] = oi.shape[0]
    else:
        gdf_setores.at[index,'qtd_clientes'] = 0
```

```python
gdf_setores.qtd_clientes.value_counts()
```

### Achata os dados para 100, para facilitar a visualizacao

```python
#gdf_setores[gdf_setores.qtd_clientes > 100].qtd_clientes = 100
gdf_setores.loc[gdf_setores['qtd_clientes'] >= 66, ['qtd_clientes']] = 10
```

```python
gdf_setores.qtd_clientes.value_counts()
```

```python
mapa_choropleth = folium.Map(location=[-15.7836, -47.7750],zoom_start=10)
```

```python
folium.Choropleth(
    geo_data=gdf_setores.to_json(),
    name='choropleth',
    data=gdf_setores,
    columns=['Name', 'qtd_clientes'],
    key_on='feature.properties.Name',
    #fill_color='YlGn',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Quantidade de clientes'
).add_to(mapa_choropleth)
```

```python
mapa_choropleth.save("qtd-clientes.html")
```
